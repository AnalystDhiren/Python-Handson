# DOUBLE LINE PLOT WITH CO-ORDINATES & Legend

```python
X = (1,4,5,6,9)
Y = (2,4,5,8,7)
X1 = (3,6,9,7,7)
Y1= (4,6,8,9,10)
plt.plot(X,Y,'r',label = "Profit")
plt.plot(X1,Y1,'b',label = "Tax")
plt.plot(X1,Y1,'ro')
plt.plot(X,Y,'go')
plt.legend()
plt.xlabel("Unites")
plt.ylabel("Count")
plt.title("Line graph")
plt.show()
```

# BAR PLOT

```python
X = (1,4,5,6,9)
Y = (2,4,5,8,7)
X1 = (3,6,9,7,7)
Y1= (4,6,8,9,10)
plt.bar(X,Y,color ='red',label = "Profit")
plt.bar(X1,Y1,color ='blue',label = "Tax")
plt.plot(X1,Y1,'ro')
plt.plot(X,Y,'go')
plt.legend()
plt.xlabel("Unites")
plt.ylabel("Count")
plt.title("Line graph")
plt.show()
```

# HISTOGRAM
