# STATISTICAL RELATIONSHIP

```python
sns.relplot(x='marks',y='hours',hue='age',data=dataset1) #hue is legend
```

```python
sns.relplot(x='hours',y='marks',kind = 'line', style ='internet', hue='age',data=dataset1) #hue is legend
```

# statistical relation multiple plot

```python
sns.relplot(x='hours',y='marks',data=dataset1,col='internet',hue="age") #hue is legend, col seperates the plots
```
